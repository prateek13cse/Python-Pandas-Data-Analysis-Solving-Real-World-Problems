The Data from a police check post is given.

# Instruction ( For Data Cleaning)

# 1- Remove the column that only contains missing values

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.drop(columns = 'country_name', inplace = True)
```

```python
df
```

# Question (Based on Filtering + Value Counts)

# 2- For Speeding, were Men or Women stopped more often?

```python
df[df.violation == 'Speeding'].driver_gender.value_counts()
```

# Question (Groupby)

# 3- Does gender affect who gets searched during a stop?

```python
df.groupby('driver_gender').search_conducted.sum()
```

```python
df.search_conducted.value_counts()
```

# Question (mapping + DataType casting)

# 4- What is the mean stop_duration?

```python
df.stop_duration.value_counts()
```

```python
df['stop_duration'] = df['stop_duration'].map({'0-15 Min' : 7.5, '16-30 Min' : 24, '30+ Min' : 45 })
```

```python
df
```

```python
df['stop_duration'].mean()
```

```python
df.groupby('violation').driver_age.describe()
```
